```python
import numpy as np
import cvxpy as cp
#definition
files = 100
T = 5000
max_cache_size = 20
x = cp.Variable(files)
utils = np.tile(np.array([1]), (1, files ))[0]
grads = []
acc_grad = np.zeros(files)
acc_r = 0
sigmas = []
ht = []
cost = [1]
action = []

#problem constraints
standard_constraints = [x>=0, x<=1, cp.sum(x)<=max_cache_size]
```

```python
def generate_cost(files, utils, action):  # this is generating r
    r = np.zeros(files) 
    requested_id = Zipf(1.2, 1, files) - 1
    requested_id = int(requested_id)
    r[requested_id] = -1
    extended_r = np.repeat(r, 1)
    grad = utils * extended_r

    if len(action) != 0:
      cost.append(cost[-1]+1-action[-1][requested_id])

    return grad, cost

def Zipf(a: np.float64, min: np.uint64, max: np.uint64, size=None):
    """
    Generate Zipf-like random variables,
    but in inclusive [min...max] interval
    """
    if min == 0:
        raise ZeroDivisionError("")
    v = np.arange(min, max + 1)  # values to sample
    p = 1.0 / np.power(v, a)  # probabilities
    p /= np.sum(p)  # normalized

    return np.random.choice(v, size=size, replace=True, p=p)
```

```python
#sigma
sum_gs = [np.zeros(files),np.zeros(files)]

def calc_sigma(sigmas, grads, pred, sumgs):
  if len(sigmas)<2:  # no grads yet
      sigmas.append(np.ones(files))  
  else:
      sigmas.append(np.sqrt(sum_gs[1]) - np.sqrt(sum_gs[0]))
  return sigmas
```

```python
def steps(x,grad,grads,files,action,acc_r,acc_grad,sigmas,sum_gs):
  pred = grad

  if len(sigmas)<2: 
    yy = cp.sum_squares(np.random.rand(files)-x)
  else:
    sum_gs[0] = sum_gs[1] 
    sum_gs[1] = sum_gs[0] + np.square(grads[-2] - pred)
    acc_r += cp.sum(sigmas[-1]/2 *(action[-1] - x)**2)
    yy = acc_r + (acc_grad + pred) @ x

  objective = cp.Minimize(yy)
  problem = cp.Problem(objective, standard_constraints)
  result = problem.solve(warm_start=True)

  #print(x.value)
  action.append(x.value)
  grads.append(grad)
  acc_grad += grads[-1]
  sigmas = calc_sigma(sigmas, sum_gs, grads, pred)
  return grads,acc_grad,sigmas,sum_gs,acc_r,action
```

```python
# main process
for i in range(1,T+1):
  print(i)
  grad, cost = generate_cost(files, utils, action)
  grads,acc_grad,sigmas,sum_gs,acc_r,action = steps(x,grad,grads,files,action,acc_r,acc_grad,sigmas,sum_gs)
  if i%100 ==0:
    np.save('/content/drive/MyDrive/data online learning/cost_pcoc.npy',np.array(cost))
```
